univariate analysis(How is distribution of gender)

```python
f,ax=plt.subplots(1,2,figsize=(20,10))
sns.countplot(x=df['gender'],data=df,palette ='bright',ax=ax[0],saturation=0.95)
for container in ax[0].containers:
    ax[0].bar_label(container,color='black',size=20)
    
plt.pie(x=df['gender'].value_counts(),labels=['Male','Female'],explode=[0,0.1],autopct='%1.1f%%',shadow=True,colors=['#ff4d4d','#ff8000'])
plt.show()
```

```python
# Group data by gender and calculate mean scores
#checking is gender has impact on student's performance?

gender_group = df.groupby('gender')[['average_score', 'math_score']].mean()
print(gender_group)

# Figure size
plt.figure(figsize=(10,8))

# X-axis labels
X = ['Average Score', 'Math Score']

# Female scores
female_scores = [
    gender_group.loc['female', 'average_score'],
    gender_group.loc['female', 'math_score']
]

# Male scores
male_scores = [
    gender_group.loc['male', 'average_score'],
    gender_group.loc['male', 'math_score']
]

# Position of bars
X_axis = np.arange(len(X))

# Male bars
plt.bar(X_axis - 0.2,
        male_scores,
        0.4,
        label='Male')

# Female bars
plt.bar(X_axis + 0.2,
        female_scores,
        0.4,
        label='Female')

# X-axis labels
plt.xticks(X_axis, X)

# Y-axis label
plt.ylabel("Marks")

# Title
plt.title("Average vs Math Scores by Gender", fontweight='bold')

# Legend
plt.legend()

# Show plot
plt.show()
```
